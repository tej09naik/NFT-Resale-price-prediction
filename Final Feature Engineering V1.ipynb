```python
import pandas as pd
import numpy as np
import json
from sklearn.linear_model import LinearRegression
```

```python
# 1. Load the dataset
data = pd.read_csv("First Draft dataset.csv")
```

```python
data.head()
```

```python
# 2. Drop unnecessary columns
drop_cols = ['__indexer_id', '__confirmed', '__block_number', 'block_number', 'log_index', 'transaction_hash', 
             'exchange_name', 'contract_version', 'aggregator_name', 'contract_address', 'is_multi_token_sale',
             'multi_token_sale_index', 'price', 'native_price', 'payment_token_address', 'quantity', 'seller_address', 
             'buyer_address', 'supply', 'name', 'description', 'image_url', 'external_url', 'media_url', 'metadata_url', 
             'last_refreshed', 'flattened_properties', '__updated_block_number']
data = data.drop(columns=drop_cols, errors='ignore')
```

```python
data.head(50)
```

```python
# 3. Extracting Properties
def extract_trait_value(properties, trait):
    for prop in properties:
        if prop['trait_type'] == trait:
            return prop['value']
    return None

properties = ['Background', 'Fur', 'Eyes', 'Clothes', 'Hat', 'Mouth', 'Earring']
for prop in properties:
    data[prop] = data['properties'].apply(lambda x: extract_trait_value(json.loads(x), prop))
```

```python
# 4. Rarity Scores
total_nfts_per_collection = data['collection_name'].value_counts()
for trait in properties:
    trait_counts = data.groupby(['collection_name', trait]).size()
    rarity_score = (1 / trait_counts).reset_index().rename({0: f'{trait}_rarity_score'}, axis=1)
    for index, row in rarity_score.iterrows():
        collection = row['collection_name']
        rarity_score.at[index, f'{trait}_rarity_score'] = row[f'{trait}_rarity_score'] * total_nfts_per_collection[collection]
    data = pd.merge(data, rarity_score, on=['collection_name', trait], how='left')
```

```python
# Sort the data by token_id, collection_name, and timestamp
data = data.sort_values(by=['token_id', 'collection_name', 'timestamp'])

# Determine the timestamp of the first sale for each NFT
data['first_sale_timestamp'] = data.groupby(['token_id', 'collection_name'])['timestamp'].transform('first')

# Update the transaction_type based on the new logic
data['transaction_type'] = np.where(data['timestamp'] == data['first_sale_timestamp'], 'sale', 'resale')

# Drop the temporary column
data.drop('first_sale_timestamp', axis=1, inplace=True)
```

```python
data.head(50)
```

```python
# 5. Transaction Type and Resale Value Calculation
#data['transaction_type'] = np.where(data['timestamp'] > data['minted_timestamp'], 'resale', 'minted')
#data['resale_value'] = data.groupby(['token_id', 'collection_name'])['usd_price'].diff().fillna(0)
#data['historical_median_price_corrected'] = data[data['transaction_type'] == 'resale'].groupby(['token_id', 'collection_name'])['usd_price'].transform('median')

# For 'sale' transactions, set resale_value_corrected to 0
data.loc[data['transaction_type'] == 'sale', 'resale_value'] = 0

# For 'resale' transactions, compute the difference from the previous transaction's usd_price
data.loc[data['transaction_type'] == 'resale', 'resale_value'] = data.groupby(['token_id', 'collection_name'])['usd_price'].diff().fillna(0)

# Historical Median Price Calculation:
#data['_median_resale_price'] = data[data['transaction_type'] == 'resale'].groupby(['token_id', 'collection_name'])['usd_price'].transform('median')
```

```python

# Convert 'timestamp' column to datetime format
data['timestamp'] = pd.to_datetime(data['timestamp'])

# Sort the data by timestamp, token_id, and collection_name
data = data.sort_values(by=['token_id', 'collection_name', 'timestamp'])

# Compute the corrected median resale price
data['median_resale_price'] = data.groupby(['token_id', 'collection_name'])['usd_price'].transform('median')
```

```python
# Assuming 'timestamp' and 'minted_timestamp' are in string format
data['timestamp'] = pd.to_datetime(data['timestamp'])
data['minted_timestamp'] = pd.to_datetime(data['minted_timestamp'])

# 6. Time Features
data['days_to_first_resale'] = (data.groupby(['token_id', 'collection_name'])['timestamp'].transform('min') - data['minted_timestamp']).dt.days
data['days_since_minting'] = (data['timestamp'] - data['minted_timestamp']).dt.days
data['transaction_frequency'] = data.groupby(['token_id', 'collection_name'])['timestamp'].transform('count')
```

```python
# 7. Price Volatility
data['price_volatility'] = data[data['transaction_type'] == 'resale'].groupby(['token_id', 'collection_name'])['usd_price'].transform('std')

```

```python
# 8. Seasonality and Time Features
data['timestamp'] = pd.to_datetime(data['timestamp'])
data['month'] = data['timestamp'].dt.strftime('%B')
data['day_of_week'] = data['timestamp'].dt.strftime('%A')
```

```python
from sklearn.linear_model import LinearRegression

# Filter the data to consider only resale transactions
resale_data_temp = data[data['transaction_type'] == 'resale'].copy()

def calculate_trend(prices):
    if len(prices) < 2:
        return 'stable'
    x = np.arange(len(prices)).reshape(-1, 1)
    model = LinearRegression().fit(x, prices)
    if model.coef_ > 0:
        return 'upward'
    elif model.coef_ < 0:
        return 'downward'
    else:
        return 'stable'

# Calculating the trend using the usd_price values of resale transactions
resale_data_temp['linear_regression_trend'] = resale_data_temp.groupby(['token_id', 'collection_name'])['usd_price'].transform(lambda x: calculate_trend(x.tolist()))

# Creating a mapping of (token_id, collection_name) to its linear_regression_trend
trend_mapping = resale_data_temp.drop_duplicates(['token_id', 'collection_name']).set_index(['token_id', 'collection_name'])['linear_regression_trend'].to_dict()

# Updating the data DataFrame directly with this trend information
data['linear_regression_trend'] = data.set_index(['token_id', 'collection_name']).index.map(trend_mapping.get)
data['linear_regression_trend'] = data['linear_regression_trend'].fillna('stable')  # Fill NaNs with 'stable'
```
